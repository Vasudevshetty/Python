```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2

fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[1].plot(x, y)

for current_ax in axes:
    current_ax.plot(x, y)
    
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(3,2), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='square of numbers')
ax.plot(x, x**3, label='cube of numbers')
fig.savefig('output.png')
ax.legend(loc=0)
```
