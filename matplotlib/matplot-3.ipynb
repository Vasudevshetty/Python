```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
x = np.linspace(0, 5, 11)
y = x * x

fig = plt.figure()
ax = fig.add_axes([0, 0, 1,1])
# plot appearance
ax.plot(x, y, color='#111', linewidth=2, linestyle='--', marker='o', markersize=8, markerfacecolor='yellow', markeredgewidth=2, 
        markeredgecolor='green')

ax.set_xlim([0, 2])
ax.set_ylim([0, 2])
```
