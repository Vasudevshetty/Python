```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2

x, y
plt.plot(x, y)
plt.xlabel('num')
plt.ylabel('squares')
plt.title('n^2 graph')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "r")

plt.subplot(1, 2, 2)
plt.plot(y, x, "b")
```

```python
# OO
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

axes.set_xlabel("num")
axes.set_title("n2")
axes.set_ylabel("squares")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 0.8, 0.8])
axes2 = fig.add_axes([0.05, 0.35, 0.4, 0.4])

axes1.plot(x, y)
axes2.plot(y, x)
```
